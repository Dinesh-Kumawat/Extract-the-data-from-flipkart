# Function

```python
import requests
from bs4 import BeautifulSoup





def get_values(searchkey = None):
    url = f"https://www.flipkart.com/search?q={searchkey}&sid=tyy%2C4io&as=on&as-show=on&otracker=AS_QueryStore_OrganicAutoSuggest_1_4_na_na_ps&otracker1=AS_QueryStore_OrganicAutoSuggest_1_4_na_na_ps&as-pos=1&as-type=RECENT&suggestionId=mobiles%7CMobiles&requestId=7a8478fc-abe2-447f-b4bf-f44d934979ba&as-searchtext=mobi"
    resp = requests.get(url) 
    if resp.status_code == 200:
        flip = BeautifulSoup(resp.text, 'html')
        name = [i.text for i in flip.select("._4rR01T")]
        rating = [i.text for i in flip.select("._3LWZlK")]
        actual_price = [i.text for i in flip.select('._3I9_wc._27UcVY')]
        price_after_discount = [i.text for i in flip.select("._30jeq3._1_WHN1")]
        m = min([len(name),len(rating),len(actual_price),len(price_after_discount)])
        for i in range(m):
            print(f"\nName --> {name[i]}".center(80))
            print(f"Rating --> {rating[i]}".center(80))
            print(f"Actual Price --> {actual_price[i]}".center(80))
            print(f"Price after Discount --> {price_after_discount[i]}".center(80))
            print("<->"*40,"\n")
    else:
        print("\n URL is Incorrect/searchkey is wrong")
```

```python
get_values("moblie under 1000")
```
